```python
import pandas as pd
from datetime import datetime
from pandas.io.data import DataReader
```

```python
goog = DataReader("GOOG", "yahoo", datetime(2000, 1, 1), datetime(2012, 1, 1))
```

```python
goog.plot()
```

```python
goog['Adj Close'].plot()
```

```python
pd.rolling_mean(goog['Adj Close'], 20).plot()
```
